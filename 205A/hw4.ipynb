```python
import numpy as np
import math
import matplotlib
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Exponential distribution with parameter = 2
\begin{align}
\theta_0 = 2
\\pdf = \theta e^{-\theta}
\end{align}


Generate a random sample of size n = 1000

```python
s = np.random.exponential(scale=0.5, size=[1000,1000]) #do it for 1000 times
theta_0 = 2
theta_0_sq = 4
cv = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence
                                df = 1)   # and degree of freedom 1
x_bar = np.mean(s, axis=1)
```

Computer LR, Wald and LM test statistics

```python
#LR Test
#According to Lecture Notes, it is not desired to use finite sample test, so we use aymptotic LR Test
LR = -2000 - 2000 * np.log(x_bar) - 2000 * np.log(theta_0) + 2000 * theta_0 * x_bar
Wald = 1000 * np.power(np.reciprocal(x_bar) - theta_0, 2) / theta_0_sq
LM = 1000 * np.power(1-  theta_0 * x_bar, 2)
r_LR = (LR-cv)>0
r_Wald = (Wald-cv)>0
r_LM = (LM-cv)>0
```

```python
r_LR.sum()/1000
```

```python
r_Wald.sum()/1000
```

```python
r_LM.sum()/1000
```
