```python
import numpy as np
import math
import matplotlib
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
n_list = [1, 5, 10, 25, 50, 100]
```

(a) Bernoulli Distribution with p = 0.78
\begin{align}
\mu = 0.78 , \ \sigma^2 = 0.78 * (1-0.78) = 0.1716
\end{align}

```python
p = 0.78
var = 0.78 * (1 - 0.78)

for n in n_list:
    s = np.random.binomial(1, p, [1000,n])
    x_bar = np.mean(s, axis=1)

    z = math.sqrt(n)*(x_bar-p)/math.sqrt(var)

    fig, ax = plt.subplots()

    n2, bins, patches = ax.hist(z, 'auto', density=1)

    sigma = 1
    mu = 0

    y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
         np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

    ax.plot(bins, y, '--')

    fig.tight_layout()
    ax.set_title(r'Histogram of Sample Size: n= %i' %n)
    plt.show()
```

(c) Exponential distribution with mean 0.5
\begin{align}
\mu = \beta = 0.5 , \ \sigma^2 = \beta^2 = 0.25
\end{align}

```python
mean = 0.5
var = 0.25
for n in n_list:
    s = np.random.exponential(scale=0.5, size=[1000, n])
    x_bar = np.mean(s, axis=1)

    z = math.sqrt(n)*(x_bar-mean)/math.sqrt(var)

    fig, ax = plt.subplots()

    n2, bins, patches = ax.hist(z, 'auto', density=1)

    sigma = 1
    mu = 0

    y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
         np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

    ax.plot(bins, y, '--')

    fig.tight_layout()
    ax.set_title(r'Histogram of Sample Size: n= %i' %n)
    plt.show()
```

(d) Distribution with the lognormal distribution with
\begin{equation}
\mu = 0 , \sigma^2 = 1
\end{equation}
We have
\begin{equation}
\\mean = exp(\mu+\frac{\sigma^2}{2}) = exp(\frac{1}{2}),
\\ Variance = [exp(\sigma^2)-1] exp( 2\mu+\sigma^2) = e^2 - e
\end{equation}

```python
mean = math.exp(1/2)
var = math.exp(2) - math.exp(1)
for n in n_list:
    s = np.random.lognormal(mean=0, sigma=1, size=[1000, n])
    x_bar = np.mean(s, axis=1)

    z = math.sqrt(n)*(x_bar-mean)/math.sqrt(var)

    fig, ax = plt.subplots()

    n2, bins, patches = ax.hist(z, 'auto', density=1)

    sigma = 1
    mu = 0

    y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
         np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

    ax.plot(bins, y, '--')

    fig.tight_layout()
    ax.set_title(r'Histogram of Sample Size: n= %i' %n)
    plt.show()
```
